```python
from IPython.display import display, JSON
import ipywidgets as widgets
import json
import requests
import re

# åˆ›å»ºè¾“å…¥æ¡†
text_area = widgets.Textarea(
    value='',
    placeholder='è¯·ç²˜è´´æ–°é—»å†…å®¹...',
    description='æ–°é—»å†…å®¹ï¼š',
    layout=widgets.Layout(width='95%', height='200px')
)

# åˆ›å»ºæŒ‰é’®
button = widgets.Button(
    description='æå–5W1H',
    button_style='primary',
    layout=widgets.Layout(width='200px')
)

# åˆ›å»ºè¾“å‡ºåŒºåŸŸ
output = widgets.Output()

# åˆ›å»ºä¸‹è½½é“¾æ¥ï¼ˆåˆå§‹ä¸ºç©ºï¼‰
download_link = widgets.HTML()

def extract_5w(news_text):
    """
    æå–æ–°é—»çš„5W1Hä¿¡æ¯
    è¿™é‡Œä½¿ç”¨æœ¬åœ°å¤„ç†æ–¹å¼ï¼Œé¿å…APIè°ƒç”¨çš„ç¼–ç é—®é¢˜
    """
    try:
        # æ¸…ç†æ–‡æœ¬
        clean_text = news_text.strip()
        if not clean_text:
            return {"error": "è¯·è¾“å…¥æ–°é—»å†…å®¹"}

        # ç®€å•çš„å…³é”®è¯æå–ï¼ˆç¤ºä¾‹ï¼‰
        # å®é™…é¡¹ç›®ä¸­ä½ å¯èƒ½éœ€è¦è°ƒç”¨AI API
        result = {
            "What": extract_what(clean_text),
            "Who": extract_who(clean_text),
            "When": extract_when(clean_text),
            "Where": extract_where(clean_text),
            "Why": extract_why(clean_text),
            "How": extract_how(clean_text),
            "åŸæ–‡": clean_text[:200] + "..." if len(clean_text) > 200 else clean_text
        }

        return result

    except Exception as e:
        return {"error": f"å¤„ç†é”™è¯¯: {str(e)}"}

def extract_what(text):
    """æå–äº‹ä»¶å†…å®¹"""
    # ç®€å•çš„äº‹ä»¶æå–é€»è¾‘
    sentences = text.split('ã€‚')
    if sentences:
        return sentences[0] + "ã€‚"
    return "æœªèƒ½è¯†åˆ«å…·ä½“äº‹ä»¶"

def extract_who(text):
    """æå–äººç‰©"""
    # ç®€å•çš„äººåè¯†åˆ«
    import re
    # æŸ¥æ‰¾å¸¸è§çš„äººåæ¨¡å¼
    names = re.findall(r'[A-Za-z\u4e00-\u9fa5]{2,4}(?:å…ˆç”Ÿ|å¥³å£«|æ€»è£|ç»ç†|ä¸»å¸­|éƒ¨é•¿|å¸‚é•¿)', text)
    if names:
        return "ã€".join(set(names[:3]))  # è¿”å›å‰3ä¸ªä¸é‡å¤çš„åå­—
    return "ç›¸å…³äººå‘˜"

def extract_when(text):
    """æå–æ—¶é—´"""
    import re
    # åŒ¹é…æ—¶é—´è¡¨è¾¾å¼
    time_patterns = [
        r'\d{4}å¹´\d{1,2}æœˆ\d{1,2}æ—¥',
        r'\d{1,2}æœˆ\d{1,2}æ—¥',
        r'ä»Šå¤©|æ˜¨å¤©|æ˜å¤©|ä»Šæ—¥|æ˜¨æ—¥|æ˜æ—¥',
        r'ä¸Šåˆ|ä¸‹åˆ|æ™šä¸Š|å‡Œæ™¨',
        r'\d{1,2}:\d{2}'
    ]

    times = []
    for pattern in time_patterns:
        matches = re.findall(pattern, text)
        times.extend(matches)

    if times:
        return "ã€".join(set(times[:2]))
    return "æ—¶é—´å¾…ç¡®è®¤"

def extract_where(text):
    """æå–åœ°ç‚¹"""
    import re
    # åŒ¹é…åœ°ç‚¹è¡¨è¾¾å¼
    place_patterns = [
        r'[A-Za-z\u4e00-\u9fa5]+(?:å¸‚|çœ|å¿|åŒº|è¡—|è·¯|æ‘|é•‡)',
        r'[A-Za-z\u4e00-\u9fa5]+(?:å¤§å­¦|å­¦æ ¡|åŒ»é™¢|å…¬å¸|é“¶è¡Œ)',
        r'åŒ—äº¬|ä¸Šæµ·|å¹¿å·|æ·±åœ³|æ­å·|å—äº¬|æ­¦æ±‰|æˆéƒ½|é‡åº†|å¤©æ´¥'
    ]

    places = []
    for pattern in place_patterns:
        matches = re.findall(pattern, text)
        places.extend(matches)

    if places:
        return "ã€".join(set(places[:3]))
    return "åœ°ç‚¹å¾…ç¡®è®¤"

def extract_why(text):
    """æå–åŸå› """
    # æŸ¥æ‰¾è¡¨ç¤ºåŸå› çš„å…³é”®è¯
    cause_keywords = ['å› ä¸º', 'ç”±äº', 'å› ', 'åŸå› æ˜¯', 'å¯¼è‡´', 'é€ æˆ']
    sentences = text.split('ã€‚')

    for sentence in sentences:
        for keyword in cause_keywords:
            if keyword in sentence:
                return sentence.strip() + "ã€‚"

    return "åŸå› éœ€è¿›ä¸€æ­¥åˆ†æ"

def extract_how(text):
    """æå–æ–¹å¼"""
    # æŸ¥æ‰¾è¡¨ç¤ºæ–¹å¼çš„å…³é”®è¯
    method_keywords = ['é€šè¿‡', 'é‡‡ç”¨', 'ä½¿ç”¨', 'åˆ©ç”¨', 'å€ŸåŠ©', 'æ–¹å¼', 'æ–¹æ³•']
    sentences = text.split('ã€‚')

    for sentence in sentences:
        for keyword in method_keywords:
            if keyword in sentence:
                return sentence.strip() + "ã€‚"

    return "æ–¹å¼å¾…åˆ†æ"

def on_button_click(b):
    """æŒ‰é’®ç‚¹å‡»äº‹ä»¶å¤„ç†"""
    with output:
        output.clear_output()

        news_text = text_area.value.strip()
        if not news_text:
            print("âŒ è¯·å…ˆè¾“å…¥æ–°é—»å†…å®¹ï¼")
            return

        print("ğŸ”„ æ­£åœ¨æå–5W1Hä¿¡æ¯...")
        result = extract_5w(news_text)

        if "error" in result:
            print(f"âŒ {result['error']}")
            return

        print("âœ… æå–å®Œæˆï¼ç»“æœå¦‚ä¸‹ï¼š")
        print("=" * 50)

        # æ˜¾ç¤ºç»“æœ
        for key, value in result.items():
            if key != "åŸæ–‡":
                print(f"ğŸ“ {key}: {value}")
                print("-" * 30)

        # ç”ŸæˆJSONä¸‹è½½
        create_download_link(result)

def create_download_link(result):
    """åˆ›å»ºä¸‹è½½é“¾æ¥"""
    json_str = json.dumps(result, ensure_ascii=False, indent=2)

    # åˆ›å»ºä¸‹è½½é“¾æ¥HTML
    download_html = f'''
    <div style="margin-top: 10px; padding: 10px; background-color: #f0f0f0; border-radius: 5px;">
        <strong>ğŸ“ ä¸‹è½½ç»“æœï¼š</strong><br>
        <textarea readonly style="width: 100%; height: 100px; margin-top: 5px;">{json_str}</textarea>
        <p style="font-size: 12px; color: #666;">æç¤ºï¼šå¤åˆ¶ä¸Šé¢çš„JSONå†…å®¹ä¿å­˜ä¸º.jsonæ–‡ä»¶</p>
    </div>
    '''
    download_link.value = download_html

# ç»‘å®šæŒ‰é’®ç‚¹å‡»äº‹ä»¶
button.on_click(on_button_click)

# æ˜¾ç¤ºç»„ä»¶
print("ğŸ“° æ–°é—»5W1Hä¿¡æ¯æå–å™¨")
print("=" * 40)
print("ä½¿ç”¨è¯´æ˜ï¼š")
print("1. åœ¨æ–‡æœ¬æ¡†ä¸­ç²˜è´´æ–°é—»å†…å®¹")
print("2. ç‚¹å‡»'æå–5W1H'æŒ‰é’®")
print("3. æŸ¥çœ‹æå–ç»“æœå¹¶ä¸‹è½½JSONæ–‡ä»¶")
print("=" * 40)

display(text_area, button, output, download_link)
```

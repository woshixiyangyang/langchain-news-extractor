```python
from IPython.display import display, JSON
import ipywidgets as widgets
import json
import os
from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.schema import HumanMessage, SystemMessage

# è®¾ç½®APIå¯†é’¥ï¼ˆè¯·æ›¿æ¢ä¸ºä½ çš„å¯†é’¥ï¼‰
# os.environ["OPENAI_API_KEY"] = "your-api-key-here"

# åˆ›å»ºè¾“å…¥æ¡†
text_area = widgets.Textarea(
    value='',
    placeholder='è¯·ç²˜è´´æ–°é—»å†…å®¹...',
    description='æ–°é—»å†…å®¹ï¼š',
    layout=widgets.Layout(width='95%', height='200px')
)

# APIé€‰æ‹©ä¸‹æ‹‰æ¡†
api_selector = widgets.Dropdown(
    options=[
        ('OpenAI GPT-3.5', 'openai-gpt35'),
        ('OpenAI GPT-4', 'openai-gpt4'),
        ('å…¶ä»–LLM', 'other')
    ],
    value='openai-gpt35',
    description='é€‰æ‹©æ¨¡å‹ï¼š',
    layout=widgets.Layout(width='300px')
)

# APIå¯†é’¥è¾“å…¥æ¡†
api_key_input = widgets.Password(
    placeholder='è¾“å…¥ä½ çš„APIå¯†é’¥...',
    description='APIå¯†é’¥ï¼š',
    layout=widgets.Layout(width='400px')
)

# åˆ›å»ºæŒ‰é’®
button = widgets.Button(
    description='ğŸ” æå–5W1H',
    button_style='primary',
    layout=widgets.Layout(width='200px')
)

# åˆ›å»ºè¾“å‡ºåŒºåŸŸ
output = widgets.Output()

# åˆ›å»ºä¸‹è½½é“¾æ¥
download_link = widgets.HTML()

def create_langchain_extractor(api_key, model_type):
    """åˆ›å»ºLangChainæå–å™¨"""
    try:
        # è®¾ç½®APIå¯†é’¥
        os.environ["OPENAI_API_KEY"] = api_key

        # é€‰æ‹©æ¨¡å‹
        if model_type == 'openai-gpt35':
            llm = ChatOpenAI(
                model_name="gpt-3.5-turbo",
                temperature=0.3,
                max_tokens=1000
            )
        elif model_type == 'openai-gpt4':
            llm = ChatOpenAI(
                model_name="gpt-4",
                temperature=0.3,
                max_tokens=1000
            )
        else:
            # å¯ä»¥æ·»åŠ å…¶ä»–æ¨¡å‹
            llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0.3)

        return llm
    except Exception as e:
        raise Exception(f"æ¨¡å‹åˆå§‹åŒ–å¤±è´¥: {str(e)}")

def create_extraction_prompt():
    """åˆ›å»ºæå–æç¤ºè¯æ¨¡æ¿"""
    template = """
ä½ æ˜¯ä¸€ä¸ªä¸“ä¸šçš„æ–°é—»åˆ†æå¸ˆã€‚è¯·ä»ä»¥ä¸‹æ–°é—»å†…å®¹ä¸­æå–5W1Hä¿¡æ¯ï¼Œå¹¶ä»¥JSONæ ¼å¼è¿”å›ã€‚

æ–°é—»å†…å®¹ï¼š
{news_text}

è¯·ä¸¥æ ¼æŒ‰ç…§ä»¥ä¸‹JSONæ ¼å¼è¿”å›ç»“æœï¼š
{{
    "Who": "æ¶‰åŠçš„äººç‰©æˆ–ç»„ç»‡",
    "What": "å‘ç”Ÿäº†ä»€ä¹ˆäº‹ä»¶",
    "When": "äº‹ä»¶å‘ç”Ÿçš„æ—¶é—´",
    "Where": "äº‹ä»¶å‘ç”Ÿçš„åœ°ç‚¹",
    "Why": "äº‹ä»¶å‘ç”Ÿçš„åŸå› æˆ–èƒŒæ™¯",
    "How": "äº‹ä»¶æ˜¯å¦‚ä½•å‘ç”Ÿçš„æˆ–é‡‡ç”¨çš„æ–¹å¼æ–¹æ³•",
    "summary": "æ–°é—»å†…å®¹çš„ç®€è¦æ€»ç»“ï¼ˆ50å­—ä»¥å†…ï¼‰"
}}

æ³¨æ„ï¼š
1. å¦‚æœæŸä¸ªä¿¡æ¯åœ¨æ–°é—»ä¸­æ²¡æœ‰æ˜ç¡®æåŠï¼Œè¯·å¡«å†™"ä¿¡æ¯ä¸æ˜ç¡®"
2. æ—¶é—´æ ¼å¼å°½é‡ç»Ÿä¸€ä¸º"YYYYå¹´MMæœˆDDæ—¥"æˆ–å…·ä½“æè¿°
3. åœ°ç‚¹è¦å…·ä½“åˆ°åŸå¸‚æˆ–åŒºåŸŸ
4. åªè¿”å›JSONæ ¼å¼ï¼Œä¸è¦æ·»åŠ å…¶ä»–è¯´æ˜æ–‡å­—
"""

    return PromptTemplate(
        input_variables=["news_text"],
        template=template
    )

def extract_5w_with_langchain(news_text, api_key, model_type):
    """ä½¿ç”¨LangChainæå–5W1Hä¿¡æ¯"""
    try:
        # æ–‡æœ¬é¢„å¤„ç†ï¼Œé¿å…ç¼–ç é—®é¢˜
        clean_text = news_text.encode('utf-8', errors='ignore').decode('utf-8').strip()

        if not clean_text:
            return {"error": "è¯·è¾“å…¥æ–°é—»å†…å®¹"}

        if not api_key:
            return {"error": "è¯·è¾“å…¥APIå¯†é’¥"}

        # åˆ›å»ºLangChainç»„ä»¶
        llm = create_langchain_extractor(api_key, model_type)
        prompt = create_extraction_prompt()

        # åˆ›å»ºé“¾
        chain = LLMChain(llm=llm, prompt=prompt)

        # æ‰§è¡Œæå–
        result = chain.run(news_text=clean_text)

        # è§£æJSONç»“æœ
        try:
            parsed_result = json.loads(result.strip())
            # æ·»åŠ åŸæ–‡ä¿¡æ¯
            parsed_result["original_text"] = clean_text[:300] + "..." if len(clean_text) > 300 else clean_text
            return parsed_result
        except json.JSONDecodeError:
            # å¦‚æœè¿”å›çš„ä¸æ˜¯æ ‡å‡†JSONï¼Œå°è¯•æå–å…³é”®ä¿¡æ¯
            return {
                "raw_result": result,
                "error": "APIè¿”å›æ ¼å¼ä¸æ˜¯æ ‡å‡†JSONï¼Œè¯·æŸ¥çœ‹raw_resultå­—æ®µ"
            }

    except Exception as e:
        return {"error": f"å¤„ç†é”™è¯¯: {str(e)}"}

def on_button_click(b):
    """æŒ‰é’®ç‚¹å‡»äº‹ä»¶å¤„ç†"""
    with output:
        output.clear_output()

        news_text = text_area.value.strip()
        api_key = api_key_input.value.strip()
        model_type = api_selector.value

        if not news_text:
            print("âŒ è¯·å…ˆè¾“å…¥æ–°é—»å†…å®¹ï¼")
            return

        if not api_key:
            print("âŒ è¯·å…ˆè¾“å…¥APIå¯†é’¥ï¼")
            return

        print("ğŸ”„ æ­£åœ¨ä½¿ç”¨LangChainåˆ†ææ–°é—»...")

        # è·å–å½“å‰é€‰ä¸­çš„æ¨¡å‹åç§°
        current_model_name = ""
        for option in api_selector.options:
            if option[1] == model_type:  # option[1]æ˜¯å€¼ï¼Œoption[0]æ˜¯æ˜¾ç¤ºåç§°
                current_model_name = option[0]
                break

        print(f"ğŸ“Š ä½¿ç”¨æ¨¡å‹: {current_model_name}")
        print("=" * 50)

        result = extract_5w_with_langchain(news_text, api_key, model_type)

        if "error" in result:
            print(f"âŒ {result['error']}")
            if "raw_result" in result:
                print("\nåŸå§‹è¿”å›ç»“æœï¼š")
                print(result["raw_result"])
            return

        print("âœ… åˆ†æå®Œæˆï¼ç»“æœå¦‚ä¸‹ï¼š")
        print("=" * 50)

        # æ˜¾ç¤ºç»“æœ
        display_order = ["Who", "What", "When", "Where", "Why", "How", "summary"]

        for key in display_order:
            if key in result:
                emoji_map = {
                    "Who": "ğŸ‘¤",
                    "What": "ğŸ“‹",
                    "When": "â°",
                    "Where": "ğŸ“",
                    "Why": "ğŸ¤”",
                    "How": "âš¡",
                    "summary": "ğŸ“"
                }
                print(f"{emoji_map.get(key, 'ğŸ“Œ')} **{key}**: {result[key]}")
                print("-" * 40)

        # ç”Ÿæˆä¸‹è½½é“¾æ¥
        create_download_link(result)

def create_download_link(result):
    """åˆ›å»ºä¸‹è½½é“¾æ¥"""
    # ç§»é™¤ä¸éœ€è¦åœ¨ä¸‹è½½æ–‡ä»¶ä¸­çš„å­—æ®µ
    download_result = {k: v for k, v in result.items() if k not in ['original_text']}

    json_str = json.dumps(download_result, ensure_ascii=False, indent=2)

    download_html = f'''
    <div style="margin-top: 15px; padding: 15px; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
                border-radius: 10px; color: white;">
        <h4 style="margin: 0 0 10px 0;">ğŸ“ æå–ç»“æœä¸‹è½½</h4>
        <textarea readonly style="width: 100%; height: 120px; margin: 10px 0; padding: 10px;
                                  border-radius: 5px; border: none; font-family: monospace;">{json_str}</textarea>
        <p style="margin: 5px 0 0 0; font-size: 12px; opacity: 0.8;">
            ğŸ’¡ æç¤ºï¼šå¤åˆ¶ä¸Šé¢çš„JSONå†…å®¹ï¼Œä¿å­˜ä¸º news_5w1h.json æ–‡ä»¶
        </p>
    </div>
    '''
    download_link.value = download_html

# ç»‘å®šæŒ‰é’®ç‚¹å‡»äº‹ä»¶
button.on_click(on_button_click)

# åˆ›å»ºç•Œé¢å¸ƒå±€
header = widgets.HTML('''
<div style="text-align: center; padding: 20px; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            color: white; border-radius: 10px; margin-bottom: 20px;">
    <h2 style="margin: 0;">ğŸ“° LangChainæ–°é—»5W1Hæå–å™¨</h2>
    <p style="margin: 10px 0 0 0; opacity: 0.9;">åŸºäºå¤§è¯­è¨€æ¨¡å‹çš„æ™ºèƒ½æ–°é—»åˆ†æå·¥å…·</p>
</div>
''')

instructions = widgets.HTML('''
<div style="background-color: #f8f9fa; padding: 15px; border-radius: 8px; margin-bottom: 15px; border-left: 4px solid #007bff;">
    <h4 style="margin: 0 0 10px 0; color: #007bff;">ğŸ“‹ ä½¿ç”¨è¯´æ˜</h4>
    <ol style="margin: 0; padding-left: 20px;">
        <li>é€‰æ‹©è¦ä½¿ç”¨çš„AIæ¨¡å‹</li>
        <li>è¾“å…¥ä½ çš„OpenAI APIå¯†é’¥</li>
        <li>åœ¨æ–‡æœ¬æ¡†ä¸­ç²˜è´´æ–°é—»å†…å®¹</li>
        <li>ç‚¹å‡»"æå–5W1H"æŒ‰é’®å¼€å§‹åˆ†æ</li>
        <li>æŸ¥çœ‹åˆ†æç»“æœå¹¶ä¸‹è½½JSONæ–‡ä»¶</li>
    </ol>
    <p style="margin: 10px 0 0 0; color: #6c757d; font-size: 12px;">
        ğŸ”’ APIå¯†é’¥ä»…ç”¨äºæœ¬æ¬¡ä¼šè¯ï¼Œä¸ä¼šè¢«ä¿å­˜
    </p>
</div>
''')

# æ˜¾ç¤ºæ‰€æœ‰ç»„ä»¶
display(
    header,
    instructions,
    api_selector,
    api_key_input,
    text_area,
    button,
    output,
    download_link
)
```
